```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd 
df = pd.read_csv("data.csv")
df.plot()
```

```python
co2 = df['CO2']
rh = df['Humidity']
temp = df['Temperature']
```

```python
co2.plot()
```

```python
rh.plot()
```

```python
temp.plot()
```

```python
e = 6.11*10**(7.5*temp/(temp+237.3))
```

```python
a = 217*(e/(temp+273.15))
```

```python
hd = (100-rh)*a/100
```

```python
import numpy as np

ax = hd.plot()
ax.legend(['Humid Deficiency'])
# ax2=ax.twinx()
ax.plot(6.8*np.ones(hd.size),'b')
ax.legend(['HD','Recommended level: HD=6.8'])
```

```python
from matplotlib.pyplot import figure
figure(figsize=(14, 6))
ax = hd.plot()
ax.legend(['HD'])
ax2=ax.twinx()
ax2.plot(temp,'b')
ax2.legend(['Temp'])
```

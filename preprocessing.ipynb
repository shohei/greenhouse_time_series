```python
def lowcut(_df, th):
    _df = _df.copy()
    n = len(_df)    
    for i in range(n):
        x = _df.iloc[i]
        if x < th:
            _df.iloc[i] = _df.iloc[i-1]    
    return _df        
```

```python
th = 330
co2_pre = lowcut(co2, 325)
```
